# Создание сигнала в комплексном представлении

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fs = 8000 # частота дискретизации
f = 200 # частота сигнала
lenght = 200 # 0.025 sec длительность

t = lenght / fs
samples = np.linspace(0, t, int(fs*t), endpoint=False)
```

```python
z = np.exp(1j * 2 * np.pi * f * samples) # комплексный сигнал через экспоненту (формула Эйлера)
```

```python
a = z.real # реальная часть
b = z.imag # мнимая часть
```

```python
plt.figure(figsize=(20, 2), dpi=80)
plt.plot(a)
plt.title('Реальная компонента')
plt.xlabel('Время (в тактах)')
plt.ylabel('Уровень')
plt.show() 
```

```python
plt.figure(figsize=(20, 2), dpi=80)
plt.plot(b)
plt.title('Мнимая компонента')
plt.xlabel('Время (в тактах)')
plt.ylabel('Уровень')
plt.show() 
```

### Мнимая компонения без i

```python
m = (a**2 + b**2)**0.5 # модуль сигнала
```

```python
plt.figure(figsize=(20, 2), dpi=80)
plt.plot(m)
plt.title('Результирующий сигнал')
plt.xlabel('Время (в тактах)')
plt.ylabel('Уровень')
plt.show() 
```

```python
ph = np.arctan(b/a) # фаза сигнала
```

```python
plt.figure(figsize=(20, 2), dpi=80)
plt.plot(ph)
plt.title('Фаза результирующего сигнала')
plt.xlabel('Время (в тактах)')
plt.ylabel('Уровень')
plt.show() 
```
